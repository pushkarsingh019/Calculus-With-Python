```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math
```

```python
x, y = sym.symbols('x y')

exp = x * sym.exp( - (x**2 + y**2))

display(Math(f'$\\text {{The Equation is : }} f(x, y) = {sym.latex(exp)}$'))

diff_x = sym.diff(exp, x, 1)
diff_y = sym.diff(exp, y, 1)

# lambdifying the functions to plot
f = sym.lambdify((x, y), exp)
df_x = sym.lambdify((x, y), diff_x)
df_y = sym.lambdify((x, y), diff_y)

xx = np.linspace(-3, 3, 21)
yy = np.linspace(-3, 3, 21)

X,Y = np.meshgrid(xx, yy)

_, axs = plt.subplots(1,3, figsize = (8, 9))

axs[0].imshow(f(X,Y), origin = 'lower', extent = [xx[0], xx[-1], yy[0], yy[-1]])
axs[0].set_title(f'$f(x, y) = {sym.latex(exp)}$')

axs[1].imshow(df_x(X,Y), origin = 'lower', extent = [xx[0], xx[-1], yy[0], yy[-1]])
axs[1].set_title(f'$\\frac{{\partial f}}{{\partial x}} = {sym.latex(diff_x)}$')


axs[2].imshow(df_y(X,Y), origin = 'lower', extent = [xx[0], xx[-1], yy[0], yy[-1]])
axs[2].set_title(f'$\\frac{{\partial f}}{{\partial y}} = {sym.latex(diff_y)}$')
```

```python
# Gradient Field with contourf and quiver
plt.contourf(X, Y, f(X, Y), levels=40, cmap='viridis')
plt.colorbar()
plt.title(f'$f(x, y) = {sym.latex(exp)}$')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.quiver(xx, yy, df_x(X, Y), df_y(X, Y))
plt.show()
```

```python
# Gradient Feild with numpy.gradient
val_x, val_y = np.gradient(f(X,Y))
plt.contourf(xx, yy, f(X, Y), levels=40, cmap='viridis')
plt.colorbar()
plt.title(f'$f(x, y) = {sym.latex(exp)}$')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.quiver(xx, yy, val_y, val_x)
plt.show()
```
