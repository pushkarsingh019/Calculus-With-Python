```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.5,3.1)
xx = np.linspace(-4,4)
y = x**2
plt.plot(xx,y)
plt.show()
print(y)
```

```python
x = np.linspace(0.5,3.1)
xx = np.linspace(-4,4)
y = np.sqrt(x)
plt.plot(xx, y)
plt.show()
print(y)
```

```python
x = np.linspace(-10,10)
y = np.sqrt(2)*x + np.cos(6)
plt.plot(x,y)
```
