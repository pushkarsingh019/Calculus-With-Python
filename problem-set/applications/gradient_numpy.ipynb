```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 2, 1001)

fx = 2*x**2 - 2*x + 4
df = 6*x - 3

plt.plot(x, fx, label='f(x)')
plt.plot(x, df, label='df')
plt.legend()
plt.xlim([-2, 2])
plt.grid()
plt.show()
```

```python
# Gradient Descent
epochs = 100
xm = np.random.choice(x)
def func(x):
    return 2*x**2 - 2*x + 4

def diff(x):
    return 6*x - 3


for iterations in range(epochs):
    d = diff(xm)
    xm = xm - d*0.01

plt.plot(x, fx, label='f(x)')
plt.plot(x, df, label='df')
plt.plot(xm, func(xm), 'o' ,color='r' , label='minima')
plt.plot(xm, diff(xm), 'o' ,color='r' , label='minima')
plt.legend()
plt.xlim([-2, 2])
plt.grid()
plt.title(f'Empirical Local Minimum : {xm}')
plt.show()
```

```python
# Gradient Descent
epochs = 100
xm = np.random.choice(x)
approximations = []
def func(x):
    return 2*x**2 - 2*x + 4

def diff(x):
    return 6*x - 3


for iterations in range(epochs):
    approximations.append(xm)
    d = diff(xm)
    xm = xm - d*0.01

approximations = np.array(approximations)
plt.plot(x, fx, label='f(x)')
plt.plot(x, df, label='df')
plt.plot(xm, func(xm), 'o' ,color='r' , label='minima')
plt.plot(xm, diff(xm), 'o' ,color='r' , label='minima')
plt.plot(approximations, func(approximations), 'o', markersize = 3, color='hotpink')
plt.plot(approximations, diff(approximations), 'o', markersize = 3, color='purple')
plt.legend()
plt.xlim([-2, 2])
plt.grid()
plt.title(f'Empirical Local Minimum : {xm}')
plt.show()
```
