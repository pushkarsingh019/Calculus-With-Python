```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sym.symbols('x')

f = sym.sin(x)*(x - 2)**2
df = sym.diff(f, x)

fx = sym.lambdify(x, f)
dff = sym.lambdify(x, df)

xx = np.linspace(0, 5*np.pi, 1001)

xlim = [0, 5*np.pi]
plt.plot(xx, fx(xx))
plt.title(f'$f(x) = {sym.latex(f)}$')
plt.xlim(xlim)
plt.grid()
plt.show()


xlim = [0, 5*np.pi]
plt.plot(xx, dff(xx))
plt.title(f'$f(x) = {sym.latex(df)}$')
plt.xlim(xlim)
plt.axhline(0, color='black', linestyle='--')
plt.grid()
plt.show()
```

```python
iterations = 40
learning_rate  = 0.001
xx = np.linspace(xlim[0], xlim[1], 1001)
xm = np.random.choice(xx)

for i in range(iterations):
    plt.plot(xm, f.subs(x, xm), 'o', markersize = 4, color='hotpink')
    d = df.subs(x, xm)
    xm = xm - d*learning_rate


fx = sym.lambdify(x, f)
dff = sym.lambdify(x, df)

xx = np.linspace(0, 5*np.pi, 1001)

xlim = [0, 5*np.pi]
plt.plot(xx, fx(xx), label='f(x)')
plt.plot(xx, dff(xx), label="f'(x)")
plt.xlim(xlim)
plt.axhline(0, color='black', linestyle='--')
plt.legend()
plt.title(f'Estimated Local Minimum : {xm}')
plt.ylim([-200, 150])
plt.grid()
plt.show()
```
