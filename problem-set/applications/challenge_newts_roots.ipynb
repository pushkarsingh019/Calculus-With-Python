```python
import sympy as sym
import numpy as np 
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
x = sym.symbols('x')

exp = 2*x**3 - 3
roots = sym.solveset(exp, x, domain = sym.S.Reals)

display(Math((f'\\text{{Real roots of }} {sym.latex(exp)} = 0 \\text{{ are }} {sym.latex(roots)}')))
print('or', sym.N(roots))
```

```python
# Exercise 2

def newtonIter(fx, x, start_value):
    df = sym.diff(fx, x)
    return start_value - ((fx.subs(x, start_value)) / df.subs(x, start_value))

x1 = newtonIter(exp, x, 1)
print(f'Estimate of root from first iteration ', x1)

x2 = newtonIter(exp, x, x1)
print(f'Estimate of root from second iteration ', sym.N(x2), 'or ', x2)
```

```python
# Exercise 3 : Visualising Newtoons Roots
approximations = []
iteration = 10
for i in range(iteration):
    xn = newtonIter(exp, x, start_value=1 if i == 0 else xn)
    approximations.append(sym.N(xn))

plt.plot(np.arange(0,iteration,1), approximations, 's-')
plt.axhline(next(iter(sym.N(roots))), color='r', linestyle='--', label='True Root')
plt.ylim([1.00, 1.175])
plt.xlabel('Iterations')
plt.ylabel('Root approximation')
plt.title('Root Approxaimations vs Iteration')
plt.legend()
plt.show()


# Iterations on function val
approximations = np.array(approximations)
fx = sym.lambdify(x, exp)
xx = np.linspace(-1,2,1001)
plt.plot(xx, fx(xx), label='f(x)')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(next(iter(sym.N(roots))), color='r', linestyle='--', label='True Root')
plt.plot(approximations, fx(approximations), 'o', color='r',markersize=8)
plt.xlim([-1, 2])
plt.ylim([-10, 10])
plt.legend()
plt.show()
```
