```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Approx(f,x,a,b):
    f = sym.sqrt(x) + 2*x
    df = sym.diff(f, x)
    approximation = (df*(b - a)).subs(x, a) + f.subs(x, a)
    real_value = f.subs(x, b)
    return approximation, real_value
```

```python
x = sym.symbols('x')

linear_approx, real = Approx(sym.sqrt(x) + 2*x, x, 4, 4.37)

print('Linear Approximation: ', linear_approx)
print('Real Value ', real)
```

```python
xx = np.arange(6,14,0.5)
exp = sym.sqrt(x) + 2*x
error_val = []

for val in xx:
    linear_approx, real = Approx(exp, x,val, 8.5)
    error = real - linear_approx
    error_val.append(error)

plt.plot(xx, error_val, 's:')
plt.xlabel('Initial Guess')
plt.ylabel('Error from "True" Value')
plt.title('Error from Linear Transformation')
plt.show()

lin = []
for l in xx:
    le,rr = Approx(exp, x, l, 8.5)
    lin.append(le)

lin = np.array(lin)
y1 = sym.lambdify(x, exp)


plt.plot(np.arange(0,14,0.1), y1(np.arange(0,14,0.1)), label='f(x)')
plt.xlim([0,14])
plt.plot(xx,lin, "s:", color='r', label='approximations')
plt.plot(8.5, y1(8.5), 'ks', color='black', label=f'x0')

plt.legend()
plt.show()
```
