```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func(x):
    if x < 0:
        return np.sin(x*np.pi)
    elif x == 0:
        return 1.5
    elif x > 0:
        return -((x-2)**2)
    else:
        print('something went wrong, x not on number line')

xx = np.linspace(-1, 2, 101)
y = np.vectorize(func)

plt.xlim([-1, 2])
plt.ylim([-4, 2])
plt.plot(xx[xx < 0], y(xx[xx < 0]))
plt.plot(xx[xx > 0], y(xx[xx > 0]))
plt.plot(0, y(0), 'o', color='red')
```

```python
# Estimating Limit as x = 0
points = np.linspace(0.0001, 0.999 * 0.4, 25) * 1/0.4

plt.plot(0 + points, y(points), 'o', color='red')
plt.plot(0 - points, y(0 - points), 'o', color='orange')
plt.xlim([-0.2, 0.2])
plt.ylim([-4, 2])
plt.plot(xx[xx < 0], y(xx[xx < 0]))
plt.plot(xx[xx > 0], y(xx[xx > 0]))
plt.plot(0, y(0), 'o', color='red')

print("Limit approach from left : ", y(0 - points)[-1])
print("Limit approach from right : ", y(points)[0])
print("Function value : ", y(0))

```

```python
# In Sympy
import sympy as sym
from sympy.abc import x
y1 = sym.sin(x * sym.pi)
y2 = 1.5
y3 = -((x-2)**2)

yf = sym.Piecewise((y1, x < 0), (y2, sym.Eq(x,0)), (y3, x > 0))

print(sym.limit(yf, x, 0, dir='-'))
print(sym.limit(yf, x, 0, dir='+'))
print(sym.limit(yf, x, 0, dir='+-'))
print(yf.subs(x, 0))
```

```python
# New Equation
fx = 3 / (1-x**2)
sym.plot(fx, (x, -2.0, 2.0), ylim=(-10,10))
```

```python
print(sym.limit(fx, x, -1, dir='-'))
print(sym.limit(fx, x, -1, dir='+'))
print(sym.limit(fx, x, -1, dir='+-'))
print(fx.subs(x, 1))
```

```python
fx1 = sym.sin(x**(-1))
sym.plot(fx1, (x, -0.5, 0.5))

print('Limit approaching from left at 0 = ', sym.limit(fx1, x, 0, dir='-'))
print('Limit approaching from Right at 0 = ', sym.limit(fx1, x, 0, dir='+'))
print('Limit approaching at 0 = ', sym.limit(fx1, x, 0, dir='+-'))
print('Function Value at 0 = ', fx1.subs(x, 0))
```
