```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display, Math
```

```python
theta = sym.symbols('theta')

fx = theta**2 * sym.exp(-(theta**2)) * sym.log(theta ** 2) * sym.sin(theta)

sym.plot(fx)
```

```python
print('Value at zero = ', fx.subs(theta, 0))
print('Limit at zero = ', sym.limit(fx, theta, 0))
```

```python
print('Domain of function:')
sym.calculus.util.continuous_domain(fx,theta,sym.Reals)
```

```python
# list limit of each part, and compute their product
parts = [theta**2 , sym.exp(-theta**2) , sym.log(theta**2) , sym.sin(theta)]
c = 1
for p in parts:
  print(f'Limit of {p} as theta->0 is {sym.limit(p,theta,0)}')
  c *= sym.limit(p,theta,0,dir='+-')

print(f'\n\nProduct of all limits is {c}')
```
