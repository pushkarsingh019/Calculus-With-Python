```python
# Zeno's Paradox Algorithm
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return np.cos(x**2)**2 + np.pi

xDomain = [-2.1, 2.1]
x = np.linspace(xDomain[0], xDomain[1], 200)

a = 1
x_temp = np.array([a - 1, a + 1])
final_lower = np.empty((0,2))
final_upper = np.empty((0,2))

for i in range(10):
    lower = np.array([x_temp[0], f(x_temp[0])])
    upper = np.array([x_temp[1], f(x_temp[1])])
    final_lower = np.append(final_lower, [lower], axis=0)
    final_upper = np.append(final_upper, [upper], axis=0)
    x_temp = (x_temp + a) / 2

for point in final_lower:
    plt.plot(point[0],point[1], 'o', color='blue', fillstyle='none', markeredgewidth = 2)


for point in final_upper:
    plt.plot(point[0], point[1], 'o', color='orange', fillstyle='none', markeredgewidth = 2)

y = np.cos(x**2)**2 + np.pi

plt.plot(x, y)
plt.xlim(xDomain)
plt.title('$f(x) = cos^2(x^2) + \\pi$')
plt.show()

```
