```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
xDomain = [0.001, 6.5]
x = np.linspace(xDomain[0], xDomain[1], 100000)
y = np.cos(np.pi*x) + np.log(x)**2

def y_func(x):
    return np.cos(np.pi*x) + np.log(x)**2


plt.xlim(xDomain)
plt.ylim([-2, 10])
plt.grid()
plt.plot(x, y)
```

```python
# Another method to underatand limits
expFact = 4
g = 1.0001 - np.linspace(1,0.00001*expFact, 10)**(1/expFact)
left  = 3 - g
right = 3 + g
limit_left = y_func(left)
limit_right = y_func(right)
```

```python
plt.plot(x, y, label='function')
plt.plot(left, limit_left, 's', label='From Left')
plt.plot(right, limit_right, 'o', label='from right')
plt.legend()
plt.xlim([2, 4])
plt.show()
```

```python
print(f'From Left -- {limit_left[0]}, From Right -- {limit_right[0]}, combined -- {y_func(3)}')
```

# In Sympy

```python
from sympy.abc import x

exp = sym.cos(x*sym.pi) + sym.log(x)**2
print(exp.subs(x, 3))
print(y_func(3))
```

```python
sym.limit(exp, x, 3, dir='+')
```

```python
# Infinite Limits
x = sym.symbols('x')
exp1 = 1 / (x-2)**2
left_limit = sym.limit(exp1, x, 2, dir='-')
right_limit = sym.limit(exp1, x, 2, dir='+')
combined_limit = sym.limit(exp1, x, 2, dir='+-')

print(left_limit, right_limit, combined_limit)

new_exp = sym.lambdify(x,exp1)

xx = np.linspace(0,4, 200)
plt.plot(xx, new_exp(xx))
```

```python
# Undefined Limits
exp2 = sym.Abs(x - 2) / (x - 2)

print(sym.limit(exp2, x, 2, dir='-'))
print(sym.limit(exp2, x, 2, dir='+'))
# print(sym.limit(exp2, x, 2, dir='+-'))

sym.plot(exp2, (x, -2,4))

```
