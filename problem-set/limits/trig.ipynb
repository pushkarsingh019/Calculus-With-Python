```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math
```

```python
phi = sym.symbols('phi')

exp = sym.sin(phi) / phi

sym.plot(exp, (phi, -100, 100), title=f'${sym.latex(exp)}$')
```

```python
g = -1 / phi
h = 1 / phi
```

```python
x = np.linspace(-100,100, 100000)
y = sym.lambdify(phi,exp)
y1 = sym.lambdify(phi, g)
y2 = sym.lambdify(phi,h)

plt.plot(x, y(x), color='black', label = 'f(x)')
plt.plot(x, y1(x), color='purple', label='h(x)')
plt.plot(x, y2(x), color = "hotpink", label='g(x)')
plt.legend()
plt.ylim([-1, 1])
plt.show()
```

```python
plt.plot(x, np.sign(y(x)), color='black', label = 'f(x)')
plt.plot(x, np.sign(y1(x)), color='purple', label='h(x)')
plt.plot(x, np.sign(y2(x)), color = "hotpink", label='g(x)')
plt.legend()
plt.show()
```

```python
display(Math(f'$ {sym.latex(sym.Limit(exp, phi, sym.oo))} = {sym.latex(sym.limit(exp, phi, sym.oo))}$'))
display(Math(f'$ {sym.latex(sym.Limit(g, phi, sym.oo))} = {sym.latex(sym.limit(g, phi, sym.oo))}$'))
display(Math(f'$ {sym.latex(sym.Limit(h, phi, sym.oo))} = {sym.latex(sym.limit(h, phi, sym.oo))}$'))
```
