```python
import sympy as sym
from sympy.geometry.util import idiff
import numpy as np
```

```python
x,y = sym.symbols('x y')
exp = x*y - 1

# Implicit Differentiation of the function.
diff = idiff(exp, y, x, 1)
diff
```

```python
# Plotting the implicit function
sym.plot_implicit(exp)

sym.plot_implicit(diff, (x, -2.5, 2.5))
```

```python
subs = sym.solve(exp, y)
subs
```

```python
diff_2 = diff.subs({y : subs[0]})
```

```python
sym.plot(diff_2)
```

## Exercise Two

```python
x,y = sym.symbols('x y')
exp = x**3*y**2 - 5*x**4

diff = sym.idiff(exp, y,x)
diff
```

```python
root = sym.sqrt(5)

new_exp = diff.subs(y, root)

new_exp
```

```python
new_exp_numeric = diff.subs(y, np.sqrt(5))
new_exp_numeric
```

```python
root = sym.sqrt(5)

new_exp = diff.subs({y : root, x : sym.pi})

new_exp
```

```python
new_exp_numeric = diff.subs({y : np.sqrt(5), x  : np.pi})
new_exp_numeric
```
