```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
x,y = sym.symbols('x y')
exp= y - 2**x

diff = sym.idiff(exp, y, x)
```

```python
expression = sym.lambdify(x, 2**x)
derivative = sym.lambdify(x, diff)
```

```python
xx = np.linspace(-2, 3, 1000)

plt.plot(xx, expression(xx), label='f(x) = 2^x')
plt.plot(xx,derivative(xx), label=f'derivative = {diff}')
plt.plot(xx,np.exp(xx),  label='e^x')
plt.legend()
plt.xlim([-2, 3])
plt.ylim([0,8])
plt.grid()
plt.show()
```
