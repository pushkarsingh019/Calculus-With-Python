```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
dx = 0.01
x = np.arange(-np.pi, np.pi, dx)
xlim = [-np.pi, np.pi]

fx = 0.1*x**4 + np.exp(-(x**2)) + np.cos(x)

df = np.diff(fx) / dx
d2f = np.diff(df) / dx
d3f = np.diff(d2f) / dx


# plotting the function
plt.plot(x, fx)
plt.title('f(x)')
plt.xlim(xlim)
plt.show()

plt.plot(x[:-1], df)
plt.title('df/dx')
plt.xlim(xlim)
plt.show()

plt.plot(x[:-2], d2f)
plt.title('d^2f/dx')
plt.xlim(xlim)
plt.show()

plt.plot(x[:-3], d3f)
plt.title('d^3f/dx')
plt.xlim(xlim)
plt.show()
```

```python
# Calculating n-derivatve using np.diff

d1x = np.diff(fx, 1) / dx
d2x = np.diff(fx, 2) / dx**2
d3x = np.diff(fx, 3) / dx**3

print(np.mean(np.abs(df - d1x)))
print(np.mean(np.abs(d2f - d2x)))
print(np.mean(np.abs(d3f - d3x)))
```

```python
# Higher order derivatives in sympy
import sympy as sym
from IPython.display import display,Math

x = sym.symbols('x')

f = 3*x**4 + sym.cos(x) + sym.exp(-(x**2))

display('First order derivative', sym.diff(f, x, 1))
display('Second order derivative', sym.diff(f, x, 2))
display('Third order derivative', sym.diff(f, x, 3))
```
