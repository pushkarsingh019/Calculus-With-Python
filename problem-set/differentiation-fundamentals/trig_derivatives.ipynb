```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dx = 0.01
x = np.arange(-1.5*np.pi, 2.0 * np.pi, dx)
y = np.cos(x)
y1 = np.diff(y) / dx
y2 = -(np.sin(x))

plt.figure(figsize=(8,5))
plt.plot(x,y,label='cos(x)',linewidth=2)
plt.plot(x[:-1],y1,label='diff(cos(x))',linewidth=2)
plt.plot(x[::20],-np.sin(x[::20]),'o',label='-sin(x)')

plt.legend()
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],labels=[r'$-\pi$',r'$-\pi/2$','0',r'$\pi/2$',r'$\pi$'])
plt.xlabel('Angle (rad.)')
plt.xlim(x[[0,-1]])
plt.show()
```

```python
dx = 0.01
x = np.arange(-1.5*np.pi, 2.0 * np.pi, dx)
y = np.sin(x)
y1 = np.diff(y) / dx
y2 = (np.cos(x[::20]))

plt.figure(figsize=(8,5))
plt.plot(x,y,label='sin(x)',linewidth=2)
plt.plot(x[:-1],y1,label='diff(sin(x))',linewidth=2)
plt.plot(x[::20],y2,'o',label='cos(x)')

plt.legend()
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],labels=[r'$-\pi$',r'$-\pi/2$','0',r'$\pi/2$',r'$\pi$'])
plt.xlabel('Angle (rad.)')
plt.xlim(x[[0,-1]])
plt.show()
```
