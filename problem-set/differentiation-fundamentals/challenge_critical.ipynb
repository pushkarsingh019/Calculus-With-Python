```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
x = np.linspace(-1.5, 1, 1001)
y = x**2 + x**3


# Plotting the function
plt.xlim([-1.5, 1])
plt.plot(x, y)
plt.show()

# computing the derivative
dydx = np.diff(y) / np.mean(np.diff(x))
# plt.ylim([-1.0, 1.0])
plt.axhline(0, color='red')
plt.plot(x[:-1],dydx)
```

```python
zero = np.where(dydx == 0)[0]
zero
```
