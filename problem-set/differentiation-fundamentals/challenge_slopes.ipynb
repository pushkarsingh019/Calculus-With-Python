```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
p1 = [-1,1]
p2 = [3,6]

m = (p2[1] - p1[1]) / (p2[0] - p1[0])

plt.plot([p1[0],p2[0]],[p1[1],p2[1]],'k')
plt.plot(p1[0],p1[1],'ro',label='p1')
plt.plot(p2[0],p2[1],'bs',label='p2')

plt.title(f'The slope of the line is m={m}')

plt.plot([-2,7],[0,0],'k--',linewidth=.3)
plt.plot([0,0],[-2,7],'k--',linewidth=.3)

plt.xlim([-2,7])
plt.ylim([-2,7])
plt.legend()
plt.gca().set_aspect('equal')
plt.show()
```

```python
# Curved Function
xDomain = [-1, 5]
n = 100
x = np.linspace(xDomain[0], xDomain[1], n)
y = x**2
def y_func(x):
    return x**2

# Plotting the line and the points
plt.xlim([-2, 6])
plt.plot(x, y)
plt.title('f(x) = x^2')
for point in x:
    plt.plot(point, point**2, 's', color='blue')
plt.show()


# Computing Slopes for Each Segment
slopes = []
for point in range(len(x) - 1):
    local_slope = (y_func(x[point+1]) - y_func(x[point])) / (x[point + 1] - x[point])
    slopes.append(local_slope)

print(f'The slopes for the {len(x) - 1} segments are : {slopes}')

# Plotting Slope Series
slope_series = []
for i in range(len(x) - 1):
    slope_serie = [x[i], slopes[i]]
    slope_series.append(slope_series)

for i in range(n - 1):
    plt.plot([x[i], x[i + 1]], [slopes[i], slopes[i]], linewidth=3)
```

```python
# Comparing Global and Local Slopes
globalSlope = (y[-1]-y[0]) / (x[-1]-x[0])
aveLocalSlopes = np.mean(slopes)

print('       Global: %g' %globalSlope)
print('Ave of locals: %g' %aveLocalSlopes)
```
