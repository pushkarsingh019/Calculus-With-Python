```python
import sympy as sym
from IPython.display import display, Math
```

```python
x = sym.symbols('x')

y = ((5/4)*x) + 9/4

sym.plot(y)

print(sym.diff(y, x))
```

```python
# For y = x^2

y = x**2

print(f'The derivate of {sym.latex(y)} is {sym.latex(sym.diff(y, x))}')
diff = sym.diff(y, x)
sym.plot(y)
```

```python
sym.plot(diff, (x, -1, 6), axis_center = [0,0])

for i in [-1, 0, 2]:
    print(f'The derivative at x = {i} is dy/dx = {diff.subs(x, i)}')
```
