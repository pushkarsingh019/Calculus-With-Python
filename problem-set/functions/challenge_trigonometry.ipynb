```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xDomain = [-(np.pi), 2*np.pi]
x = np.linspace(xDomain[0], xDomain[1], 101)
f1 = np.sin(x)
f2 = (np.sin(x))**2
f3 = np.sin(x**2)

plt.grid()
plt.xlim(xDomain)
plt.plot(x,f1,label='$f(\\theta) = sin(\\theta)$')
plt.plot(x,f2,label='$f(\\theta) = sin^2(\\theta)$')
plt.plot(x,f3,label='$f(\\theta) = sin(\\theta^2)$')
plt.ylabel('$y = f(\\theta)$')
plt.xlabel('$\\text {Angle} (\\theta)$')
plt.legend()
plt.show()
```

```python
f1 = np.sin(np.cos(x))
f2 = np.cos(np.sin(x))
f3 = np.cos(x)

plt.grid()
plt.xlim(xDomain)
plt.plot(x,f1,label='$f(\\theta) = sin(cos(\\theta))$')
plt.plot(x,f2,label='$f(\\theta) = cos(sin(\\theta))$')
plt.plot(x,f3,label='$f(\\theta) = cos(\\theta))$')
plt.ylabel('$y = f(\\theta)$')
plt.xlabel('$\\text {Angle} (\\theta)$')
plt.legend()
plt.show()
```

```python
# Resolution Tan
xDomain = [0, 2*np.pi]
x1 = np.linspace(xDomain[0], xDomain[1],100)
x2 = np.linspace(xDomain[0], xDomain[1],1000)
x3 = np.linspace(xDomain[0], xDomain[1],10000)
f1 = np.tan(x1)
f2 = np.tan(x2)
f3 = np.tan(x3)

plt.plot(x1,f1, label='100 points')
plt.plot(x2,f2, label='1000 points')
plt.plot(x3,f3, label='10000 points')
plt.legend()
plt.xlim([1.55, 1.60])
```
