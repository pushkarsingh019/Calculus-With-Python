```python
import numpy as np

def random_func(a, piece):
    result = []
    if piece == 1:
        for val in a:
            choice = np.random.choice([val, val**2, val**3])
            result.append(choice)
    elif piece == 2:
        for val in a:
            if val <= 0:
                raise ValueError("All elements in 'a' must be positive for logarithm and square root.")
            choice = np.random.choice([np.log(val), np.sqrt(val)])
            result.append(choice)
    else:
        raise ValueError('Wrong option chosen')
    print(choice)
    return np.array(result)

a1 = np.linspace(-3, 0, 101)
a2 = np.linspace(0.001, 3, 101)

# Call the function with valid ranges
y1 = random_func(a1, 1)
y2 = random_func(a2, 2)

plt.xlim(-3,3)
plt.plot(a1,y1)
plt.plot(a2,y2)
plt.show()
```
