```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display, Math
```

```python
n = np.arange(1,11)
e = (1 + (1/n))**n

for i in n:
    print(f'n: {i:2.0f}, est.e: {e[i-1]:.5f}, dif to e. {np.e - e[i-1]:.5f}')
```

```python
# Visualising Difference to e.
x = np.arange(1,1001)
e = (1 + (1/x))**x
diff = np.e - e
plt.title('Difference between estimated and theoretical E.')
plt.ylabel('Difference. (theoretical - estimated)')
plt.xlabel('n')
plt.plot(x,diff)
```

```python
# Exploring E.
x = np.linspace(-2,2,100)
y1 = np.exp(x)
y2 = np.exp(x**2)
y3 = np.exp((-x)**2)
y4 = np.exp(-(x**2))
y5 = (np.exp(x)) ** 2



plt.grid()
plt.ylim([-1, 10])
plt.xlim([-2,2])
plt.plot(x,y1, color='blue', label='$ y = e^x $')
plt.plot(x,y2, color='orange', label='$ y = e^{x^2} $')
plt.plot(x,y3, '--' ,color='green', label='$ y = e^{(-x)^2} $')
plt.plot(x,y4,color='red', label='$y = e^{-(x^2)}$')
plt.plot(x,y5, color='purple', label='$ y = (e^x)^2 $')
plt.legend()
plt.show()
```

```python
# exploring e using sympy
beta = sym.var('beta')
exp = sym.exp(beta) - sym.log(beta) - np.e

sym.plot(exp, (beta, -2,2), title=f'$f(\\beta) = {sym.latex(exp)}$')
```

```python
xDomain = [-4,4]
x = np.linspace(xDomain[0], xDomain[1], 100)

y1 = np.log(x)
y2 = np.exp(x)
y3 = np.log(np.exp(x))
y4 = np.exp(np.log(x))

plt.grid()
plt.plot(x,y1,color='b', label = '$ log(x) $')
plt.plot(x,y2, color='orange', label='$ exp (x) $')
plt.plot(x,y3, color='green', label='$log(exp(x))$')
plt.plot(x,y4, 'o', color='r', label='$exp(log(x))$')
plt.xlim(xDomain)
plt.ylim([-4,4])
plt.legend()
plt.show()
```
