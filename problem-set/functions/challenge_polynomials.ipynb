```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display, Math
```

```python
# creating a third order polynomial, so four coefficients
xDomain = [-3,3]
x = np.linspace(xDomain[0], xDomain[1])
a1,a2,a3,a4 = np.random.rand(4)
y = a1 + a2*x + a3*x**2 + a4*x**3
plt.title("Polynomial Function")
plt.xlabel('$x$')
plt.ylabel('$(x)$')
plt.grid()
plt.xlim(xDomain[0], xDomain[1])
plt.plot(x,y, label=f'$y = {round(a1,2)} + {round(a2,2)}x + {round(a3,2)}x^2 + {round(a4,2)}x^3$')
plt.legend()
```

```python
# we have to simulate the following function
display(Math('$$f(x) = \\frac{x^2 - 2x}{x^2 - 4}$$'))

x = sym.symbols('x')
exp = (x**2 - 2*x) / (x**2 - 4)

sym.plot(exp, (x,xDomain[0],xDomain[1]),xlabel='x',ylim=[-10,10],title=f'$y = {sym.latex(exp)}$')
```

```python
# above function with numpy
fx = sym.lambdify(x,exp)
xx = np.linspace(-3,3)
yy = fx(xx)
plt.grid()
plt.xlim(-3,3)
plt.ylim(-10,10)
plt.plot(xx,yy)
```

```python
# exercise three: maclaurin series
order = 10

x = np.linspace(-2*np.pi,2*np.pi,100)
z = np.zeros(len(x))

for n in range(1,order+1):
    mul = (-1)**(n+1)
    top = x**(2*n-1)
    deno = np.math.factorial(2*n-1)
    fx = mul*(top / deno)
    z += fx
    plt.ylim([-5,5])
    plt.xlim(x[[0,-1]])
    plt.plot(fx)

plt.plot(x,z,'k',linewidth=3,label=f'Sum over {order} terms')
plt.plot(x[::5],np.sin(x[::5]),'bo',markerfacecolor='w',linewidth=3,label='sin(x)')
plt.ylim([-5,5])
plt.xlim(x[[0,-1]])
plt.legend()
plt.show()
```
