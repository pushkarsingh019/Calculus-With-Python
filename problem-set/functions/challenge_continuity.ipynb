```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
xDomain = [-1,2]
x1 = np.linspace(xDomain[0],0, 100)
x2 = 0
x3 = np.linspace(0.0000001,xDomain[1], 100)
y1 = np.sin(x1*np.pi)
y2 = 1.5
y3 = -((x3 - 2)**2)

plt.plot(x1,y1,label='$sin(x\\pi)$')
plt.plot(x2,y2,'o')
plt.plot(x3,y3)

plt.xlim(xDomain)
plt.legend()
```

```python
x = sym.symbols('x')
y1 = sym.sin(x*sym.pi)
y2 = 1.5
y3 = -((x - 2)**2)
y = sym.Piecewise((y1,x < 0), (y2, sym.Eq(x,0)), (y3, x > 0))

sym.plot(y, (x,xDomain[0],xDomain[1]))
```

```python
y.subs(x,0.0)
```

```python
# removable discountinuity
x = np.linspace(-1,2,80)
y = np.sin(np.pi*x) + x*2
y[np.argmin(np.abs(x - 0))] = np.pi
plt.plot(x, y, 'o')
```

```python
# Infinite discontinuity
xDomain = [-2,2]
xx = np.linspace(-2,2,1000)
y = (3) / (1 - xx**2)

plt.xlim(xDomain)
plt.ylim(-20,20)
plt.plot(xx,y)
```
