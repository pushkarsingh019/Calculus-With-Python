```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
xDomain = [-3,5]
x = np.linspace(xDomain[0], xDomain[1], 50)

piece1 = 0* (x<0)
piece2 = -2*x * ((x >=0) & (x < 3))
piece3 = x**3*(0.1) * (x>3)

y = piece1 + piece2 + piece3
plt.xlim(xDomain)
plt.plot(x,y)
```

```python
xDomain = [-3,5]
x1 = np.linspace(xDomain[0], 0, 50)
x2 = np.linspace(0, 2.99999999, 50)
x3 = np.linspace(3.1, xDomain[1], 50)

piece1 = 0* (x1<0)
piece2 = -2*x2 * ((x2 >=0) & (x2 < 3))
piece3 = x3**3*(0.1) * (x3>3)

y1 = piece1
y2 = piece2
y3 = piece3

plt.plot(x1,y1, label='Piece 1', color='blue')
plt.plot(x2,y2, label='Piece 2', color='orange')
plt.plot(x3,y3, label='Piece 3', color='green')

plt.xlim(xDomain)
plt.show()
```

```python
# in sympy
from sympy.abc import x,y
f1 = 0
f2 = -2*x
f3 = 0.1*x**3

p = sym.Piecewise((f1, x < 0), (f2, sym.And(x >= 0,x < 3)), (f3, x >= 3))

sym.plot(p, (x, xDomain[0], xDomain[1]),xlabel='x', ylabel='y', title=f'Piecewise function')
```
