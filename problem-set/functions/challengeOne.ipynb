```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as smp
```

```python
xDomain = [-2,2]
x = np.linspace(xDomain[0],xDomain[1],51)
y = np.power(x, 2) + 3*np.power(x, 3)- np.power(x, 4)
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.plot(x,y,label='$y = x^2 + 2x^3 - x^4$')
plt.legend()
plt.grid()
plt.xlim(xDomain[0],xDomain[1])
plt.show()
```

```python
# Exercise 2
import sympy as sym
b = sym.symbols('beta')
exp = -b**4 + 3*b**3 + b**2

sym.plot(exp, (b, xDomain[0],xDomain[1]), xlabel = "x", ylabel=None,title=f'$f(\\beta) = {sym.latex(exp)}$')

```

## Sympy to Numpy

```python
fx = sym.lambdify(b,exp)
xx = np.linspace(-2,2)
yy = fx(xx)

plt.plot(xx, yy)
plt.grid()
plt.xlim(xDomain[0],xDomain[1])
plt.show()
```
