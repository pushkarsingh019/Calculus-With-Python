```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
x = np.linspace(-5,5,100)
f = 2*x**2 - 4
g = 7*np.abs(x) + 3
fg = 7*np.abs(f) + 3
gf = 2*(g**2) - 4

plt.plot(x,f, label='f(x)', color='b')
plt.plot(x,g, label='g(x)', color='orange')
plt.plot(x, gf, label='f(g(x))', color='green')
plt.plot(x, fg, label='g(f(x))', color='red', linestyle='--')
plt.legend()
plt.xlim(-5,5)
plt.ylim(-10,50)
plt.plot()
```

```python
x = np.linspace(0.00001, 5, 101)

f = np.log(2*x)
g = np.exp(x) / 2
fog = np.log(2*(g))
gof = np.exp(f) / 2


plt.xlim(0.00001, 5)
plt.grid()
plt.plot(x, f, label = "f(x)", color='blue')
plt.plot(x, g, label = "g(x)", color='orange')
plt.plot(x, fog, label = "f(g(x))", color='green')
plt.plot(x, gof, label = "g(f(x))", color='red', linestyle='--')
plt.ylim([-5, 10])
plt.legend()
plt.show()
```

```python
x = sym.symbols('x')
exp1 = sym.sin(x)
exp2 = sym.log(x)
exp3 = 2*(x**2) + 5

exp = exp1.subs(x,exp2.subs(x, exp3))
sym.plot(exp, (x, -100,100))
```

```python
# Inverse function
y = sym.symbols('y')
fun = 2*x + 3

sym.solve(y - fun, x)[0]
invfun = sym.solve(y-fun, x)[0]

fun.subs(x, invfun.subs(y,4))
invfun.subs(y,fun.subs(x,4))
```
