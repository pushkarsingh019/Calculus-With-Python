```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
# partial derivatives using sympy
x, y = sym.symbols('x y')

exp = ((x**2) / (sym.pi)**3) + sym.sin(y)

diff_x = sym.diff(exp, x)
diff_y = sym.diff(exp, y)

display(Math(f'$\\frac{{\partial f}}{{\partial x}} = {sym.latex(diff_x)}$'))

display(Math(f'$\\frac{{\partial f}}{{\partial y}} = {sym.latex(diff_y)}$'))
```

```python
func = sym.lambdify((x, y), exp)
df_x = sym.lambdify((x, y), diff_x)
df_y = sym.lambdify((x, y), diff_y)

xx = np.linspace(0, 2*np.pi, 21)

X, Y = np.meshgrid(xx, xx)

_, axs = plt.subplots(1,3, figsize = (8, 9))

axs[0].imshow(func(X,Y), origin = 'lower', extent = [xx[0], xx[-1], xx[0], xx[-1]])
axs[0].set_title(f'$f(x, y) = {sym.latex(exp)}$')

axs[1].imshow(df_x(X,Y), origin = 'lower', extent = [xx[0], xx[-1], xx[0], xx[-1]])
axs[1].set_title(f'$\\frac{{\partial f}}{{\partial x}} = {sym.latex(diff_x)}$')


axs[2].imshow(df_y(X,Y), origin = 'lower', extent = [xx[0], xx[-1], xx[0], xx[-1]])
axs[2].set_title(f'$\\frac{{\partial f}}{{\partial y}} = {sym.latex(diff_y)}$')
```
